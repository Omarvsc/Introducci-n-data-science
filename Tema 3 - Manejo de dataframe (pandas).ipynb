Ejercicio 1:

Vamos a usar lo aprendido en el tema anterior para usar un bucle y guardar en un dataframe los maximos de cada variable.

```python
# Se guarda los nombres de cada variable del dataframe en un vector
col_names= base_wine_red.columns.to_list()
# Se crea un dataframe vacío
df_1=pd.DataFrame(columns={'variable','maximo'})

for i in col_names:
    #Se calcula el máximo para cada variable
    num_max=base_wine_red[i].max()
    # se guarda cada nuevo registro en una variable temporal
    row=pd.DataFrame([(i,num_max)],columns={'variable','maximo'})
    # se inserta al dataframe creado al principio cada registro "row"
    df_1=pd.concat([df_1,row],axis=0)

#Reseteamos el index
df_1.reset_index(inplace=True)    

df_1=df_1.drop(['index'],axis=1)    
```

```python
df_1
```

**Correlación**

correlación de pearson

$r = cov(x,y)/(\sigma(x)*\sigma(y)$)

Como ejemplo practico imprimiremos las correlaciones de cada variable respecto al target (quality)

```python
var=base_wine_red_orig.columns.to_list()
var.remove('quality')
target='quality'

for i in var:
    cor=base_wine_red_orig[[i,target]].corr()
    cor_2=cor[target][0]
    print("La correlación entre "+i+" y "+target+ " es: ",np.round(cor_2,2))
```

```python
base_wine_red_orig[['alcohol','quality']].plot(x='alcohol',y='quality',kind='scatter')
```

Correlacion de spearman

Correlación que se basa en rangos, por ellos se menos sensible a los valores atípicos.

```python
var=base_wine_red_orig.columns.to_list()
var.remove('quality')
target='quality'

for i in var:
    cor=base_wine_red_orig[[i,target]].corr(method='spearman')
    cor_2=cor[target][0]
    print("La correlación entre "+i+" y "+target+ " es: ",np.round(cor_2,2))
```

### Métodos conocidos para normalizar variables

Se busca estandarizar/normalizar las variables para evitar el sobreajuste y computacionalmente sea más rapido la ejecución del algoritmo. Considerar que hay modelos que no necesitan tal normalización, pero se suele realizar dependiendo de la complejidad del modelo.

**z- standardization**

<font size="5"> $z = \frac{(x - \mu)}{\sigma}$ </font>

```python
def z_standardization(df):
    return df.apply(lambda x: (x-x.mean())/x.std(), axis=0)
```

```python
base_wine_red_orig['alcohol'].describe()
```

```python
pH_z=base_wine_red_orig['alcohol']
pH_z=pd.DataFrame(pH_z)
base_wine_red_orig['alcohol_z']=z_standardization(pH_z)
```

```python
base_wine_red_orig['alcohol_z'].describe()
```

```python
base_wine_red_orig['alcohol'].hist()
base_wine_red_orig['alcohol_z'].hist()
```
